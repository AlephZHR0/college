```python
# comando para importar arquivo CSV: read_csv e salvar os dados no dataframe casa
casa = pd.read_csv('dados-carregamento-ESEG.txt')
```

```python
# mostrando as primeiras linhas do dataframe casa. Este dataframe contém informações sobre casas nos EUA e seus valores
casa.head()
```

```python
# o comando shape mostra a quantidade de linhas e colunas do dataframe
casa.shape
```

```python
# Podemos usar a função  describe() para ter uma visualização rápida  das principais estatísticas dos dados
casa.describe().T
```

```python
# mostrando informações do dataframe, ou seja, do conjunto de dados importado
# Pode usar a função info() nos informa os tipos de variáveis que temos e já é possível saber se há valores faltantes
casa.info()
```

```python
# Para visualizar um resumo das variáveis que possui dados faltantes, 
# podemos usar a função isnull() (ou a função isna() )e a função sum().
casa.isnull().sum()
```

```python
# Caso sejam poucos dados e eles não afetem os demais dados, 
# podemos fazer a exclusão das linhas com dados faltantes com a função dropna()
# e o subconjunto com as variáveis que queremos excluir.
casa.dropna(subset=['banheiros','ano-construcao'], inplace=True)
casa.isna().sum()
```

```python
# verificando o resultado da substituição
casa.head()
```

```python
# Outro problema recorrente é a duplicidade dos dados. 
# Para verificar a existência de duplicidade, usamos a função duplicated().

casa[casa.duplicated()]
```

```python
# Para eliminar os dados duplicados podemos utilizar a função drop_duplicates().
casa.drop_duplicates(inplace=True)
```

```python
# Verificando a remoção das duplicatas:

casa[casa.duplicated()]
```
