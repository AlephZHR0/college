```python
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
with open("./academia.csv", "r") as academia_file:
    uploaded =  pd.read_csv(academia_file)
```

```python
uploaded.head()
```

```python
uploaded.SEXO.value_counts().plot(kind='pie',autopct="%.2f%%",)
```

```python
sns.countplot(x="SEXO", data=uploaded)
```

```python
table = pd.crosstab(uploaded["SEXO"], uploaded["ALUNO"])
table.plot(kind="bar", stacked=True, figsize=(8,7))
```

```python
table_perc = table.div(table.sum(axis=1), axis=0)
table_perc.plot(kind="bar", stacked=True, figsize=(8,7))
```

```python
sns.histplot(uploaded["ALTURA"], kde=True, bins=2)
```

```python
%matplotlib inline
sns.boxplot(y="ALTURA", data = uploaded, palette="rainbow")
```

```python
%matplotlib inline
sns.boxplot(x="SEXO", y="ALTURA", data=uploaded, palette="rainbow")
```
