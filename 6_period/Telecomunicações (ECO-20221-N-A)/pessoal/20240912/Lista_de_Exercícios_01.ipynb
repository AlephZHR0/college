# 9. Para amostrar um sinal de v√≠deo com $f_{\text{max}} = 4\ MHz$, pelo teorema de Nyquist

## 1. Qual o menor valor da frequ√™ncia de amostragem?

Pelo Teorema de Nyquist, a frequ√™ncia de amostragem m√≠nima (fa) deve ser o dobro da maior frequ√™ncia do sinal (fmax):

```python
maior_frequencia_sinal = 4  # MHz
frequencia_amostragem = 2 * maior_frequencia_sinal # MHz
print(f'O sinal deve ser amostrado com frequ√™ncia de {frequencia_amostragem}MHz')
```

## 2. Qual √© a Banda de Guarda adotada?

A banda de guarda (BG) √© a diferen√ßa entre a frequ√™ncia de amostragem (fa) e a maior frequ√™ncia do sinal (fmax):

```python
banda_guarda = frequencia_amostragem - maior_frequencia_sinal
print(f'A banda de guarda √© de {banda_guarda}MHz')
```

## 3. Sendo a quantiza√ß√£o de $8$ bits por amostra do sinal, qual ser√° o valor da taxa de transmiss√£o?

Taxa de transmiss√£o = Frequ√™ncia de amostragem * N√∫mero de bits por amostra

```python
taxa_de_transmiss√£o = frequencia_amostragem * 8 # Mbits/s
print(f'A taxa de transmiss√£o √© de {taxa_de_transmiss√£o}Mbits/s')
```

## 4. Desenhe as curvas de amostragem.

```python
# Dados
print(f'Banda de guarda = {banda_guarda}') # Banda de guarda (MHz)
frequencia_central = 2 * maior_frequencia_sinal  # MHz
print(f'Maior frequ√™ncia do sinal = {maior_frequencia_sinal}') # Maior frequ√™ncia do sinal (MHz)
print(f'Frequ√™ncia central = {frequencia_central}') # Frequ√™ncia central (MHz)
```

# 10. Utilizando como base o programa em Python mostrado em sala. Altere suas instru√ß√µes para obter a s√©rie de Fourier das ondas a seguir. Tenham aten√ß√£o √†s equa√ß√µes integrais do sinal. O software dever√° imprimir o gr√°fico no dom√≠nio do tempo e da frequ√™ncia (Transformada de Fourier).

```python
T = 2 * np.pi  # Per√≠odo da onda
omega0 = 2 * np.pi / T  # Frequ√™ncia angular fundamental
t = np.linspace(-2*T, 2*T, 2000)  # Vetor de tempo abrangendo quatro per√≠odos
N = 50  # Aumentar para melhorar a aproxima√ß√£o
```

## 1. Onda dente de serra


```python
soma_dente_serra = np.zeros_like(t)

# A s√©rie de Fourier da onda dente de serra cont√©m apenas termos senoidais (bn)
# bn = (2 / (n * œÄ)) * (-1)^(n + 1)

for n in range(1, N + 1):
    bn = (2 / (n * np.pi)) * (-1) ** (n + 1)
    soma_dente_serra += bn * np.sin(n * omega0 * t)

# Plotar a onda dente de serra original e sua aproxima√ß√£o pela s√©rie de Fourier
plt.figure(figsize=(12, 6))
plt.plot(t, (2 / T) * (((t + T/2) % T) - T/2), label='Onda Dente de Serra Original', color='blue')
plt.plot(t, soma_dente_serra, label='Aproxima√ß√£o da S√©rie de Fourier (N={})'.format(N), color='red', linestyle='--')
plt.title('Onda Dente de Serra e sua Aproxima√ß√£o pela S√©rie de Fourier')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.xlim(-2*T, 2*T)
plt.show()

# Calcular e plotar o espectro de frequ√™ncia (magnitude dos coeficientes de Fourier)
n = np.arange(1, N + 1)  # N√∫meros harm√¥nicos
bn = (2 / (n * np.pi)) * (-1) ** (n + 1)  # Coeficientes de Fourier

plt.figure(figsize=(10, 6))
plt.stem(n * omega0, np.abs(bn))
plt.title('Espectro de Frequ√™ncia da Onda Dente de Serra')
plt.xlabel('Frequ√™ncia (rad/s)')
plt.ylabel('Magnitude dos Coeficientes de Fourier')
plt.grid(True)
plt.show()
```

## 2. Onda triangular


```python
soma_triangular = np.zeros_like(t)

# A s√©rie de Fourier da onda triangular cont√©m apenas termos cossenoidais (an) para n √≠mpar
# an = (8 / (n^2 * œÄ^2)) * (-1)^{(n - 1)/2}

for n in range(1, 2 * N, 2):  # n assume valores √≠mpares
    an = (8 / (n ** 2 * np.pi ** 2)) * (-1) ** ((n - 1) / 2)
    soma_triangular += an * np.cos(n * omega0 * t)

# Plotar a onda triangular original e sua aproxima√ß√£o pela s√©rie de Fourier
plt.figure(figsize=(12, 6))
plt.plot(t, (4 / T) * np.abs(np.mod(t, T) - T/2) - 1, label='Onda Triangular Original', color='blue')
plt.plot(t, soma_triangular, label='Aproxima√ß√£o da S√©rie de Fourier (N={})'.format(N), color='red', linestyle='--')
plt.title('Onda Triangular e sua Aproxima√ß√£o pela S√©rie de Fourier')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.xlim(-2*T, 2*T)
plt.show()

# Calcular e plotar o espectro de frequ√™ncia (magnitude dos coeficientes de Fourier)
n = np.arange(1, 2 * N, 2)  # Harm√¥nicos √≠mpares
an = (8 / (n ** 2 * np.pi ** 2)) * (-1) ** ((n - 1) / 2)  # Coeficientes de Fourier

plt.figure(figsize=(10, 6))
plt.stem(n * omega0, np.abs(an))
plt.title('Espectro de Frequ√™ncia da Onda Triangular')
plt.xlabel('Frequ√™ncia (rad/s)')
plt.ylabel('Magnitude dos Coeficientes de Fourier')
plt.grid(True)
plt.show()
```
