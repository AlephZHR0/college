```python
import pandas as pd
data:dict = {
    'Atividade': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Precedência': ['-', '-', 'A', 'B', 'C,D', 'A', 'E,F', 'C,D', 'B', 'G,H,I'],
    'Otimista': [1, 1, 5, 2, 3, 9, 1, 4, 1, 1.5],
    'Mais provável': [2.5, 3, 7, 4, 4.5, 14, 1, 11, 2, 2.5],
    'Pessimista': [7, 11, 15, 12, 9, 25, 1, 12, 9, 12.5],
}
data['Duração'] = [(data['Otimista'][i] + 4*data['Mais provável'][i] + data['Pessimista'][i])/6 for i in range(len(data['Atividade']))]
# data['Folga Total'] = 

df_1 = pd.DataFrame(data)
df_1['Folga Total'] = [0, 2, 0, 2, 4, 2, 2, 0, 14, 0]
df_1['Folga Livre'] = [0, 0, 0, 2, 2, 0, 2, 0, 14, 0]
```

```python
print('\t'.join(list(df_1.keys())))
for i, row in df_1.iterrows():
    for key in row.keys():
        print(row[key], end='\t')
    print()
```
