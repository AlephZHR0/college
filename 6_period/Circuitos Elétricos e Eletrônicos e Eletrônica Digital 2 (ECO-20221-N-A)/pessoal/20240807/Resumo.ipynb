```python
import math
```

# Representao Algbrica/retangular

**$Z = a + b*j$**

$a$ = Parte Real

$b$ = Parte Imaginria

$j$ = unidade imaginria

# Forma Trigonomtrica

**$Z = r(cos() + j * sen())$**

$r$ = Distncia do nmero complexo at a origem

$\theta$ = ngulo formado entre o nmero complexo e o eixo real

# Forma Angular/Polar

**$Z = r * \angle \theta$**

$r$ = Distncia do nmero complexo at a origem

$\angle$ = Indica que estamos utilizando a forma angular e que o valor seguinte representa o ngulo.

$\theta$ = ngulo formado entre o nmero complexo e o eixo real

# Como calcular $r$ e $\theta$

$r = |Z| = \sqrt{a^{2} + b^{2}}$

$\theta = arctan(\frac{b}{a})$

b = parte imaginria

a = parte real

```python
a = 4+3j # Nmero complexo
theta = math.atan(a.imag/a.real) # ngulo em radianos
graus = math.degrees(theta) # ngulo em graus
A = math.sqrt(a.real**2 + a.imag**2) # Mdulo/Amplitude/Comprimento
print(f'Forma polar: {A}{graus}') # Forma polar
print(f'Forma retangular: {a.real} + {a.imag}j') # Forma retangular
print(f'Forma trigonomtrica: {A} (cos({theta}) + sen({theta}))j = {A * (math.cos(theta) + A * math.sin(theta))}j') # Forma trigonomtrica
```
