```python
import math
```

# Representação Algébrica/retangular

**$Z = a + b*j$**

$a$ = Parte Real

$b$ = Parte Imaginária

$j$ = unidade imaginária

# Forma Trigonométrica

**$Z = r(cos(θ) + j * sen(θ))$**

$r$ = Distância do número complexo até a origem

$\theta$ = Ângulo formado entre o número complexo e o eixo real

# Forma Angular/Polar

**$Z = r * \angle \theta$**

$r$ = Distância do número complexo até a origem

$\angle$ = Indica que estamos utilizando a forma angular e que o valor seguinte representa o ângulo.

$\theta$ = Ângulo formado entre o número complexo e o eixo real

# Como calcular $r$ e $\theta$

$r = |Z| = \sqrt{a^{2} + b^{2}}$

$\theta = arctan(\frac{b}{a})$

b = parte imaginária

a = parte real

```python
a = 4+3j # Número complexo
theta = math.atan(a.imag/a.real) # Ângulo em radianos
graus = math.degrees(theta) # Ângulo em graus
A = math.sqrt(a.real**2 + a.imag**2) # Módulo/Amplitude/Comprimento
print(f'Forma polar: {A}∠{graus}°') # Forma polar
print(f'Forma retangular: {a.real} + {a.imag}j') # Forma retangular
print(f'Forma trigonométrica: {A} (cos({theta}) + sen({theta}))j = {A * (math.cos(theta) + A * math.sin(theta))}j') # Forma trigonométrica
```
